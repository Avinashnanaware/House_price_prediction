```python
#miss_val_per=df.isnull().sum()/df.shape[0]*100
miss_val_per=df.isnull().mean()
miss_val_per
```

```python
miss_val_col_gre_20=miss_val_per[miss_val_per> 20].keys()
miss_val_col_gre_20
```

```python
df2_drop_col=df.drop(columns=miss_val_col_gre_20)
```

```python
df2_drop_col.shape
```

```python
df3_num=df2_drop_col.select_dtypes(include=['int64','float64'])
df3_num.head()
```

```python
sns.heatmap(df3_num.isnull())
```

```python
df3_num_isnull=df3_num.isnull().sum()
df3_num_isnull
```

```python
mising_num=[var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
mising_num
#miss_num=df3_num_isnull[df3_num_isnull> 0]
#miss_num
```

```python
sns.set()
for i, var in enumerate(mising_num):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':5, 'color':'#DC143C'})
```

```python
df4_mean=df3_num.fillna(df3_num.mean())
```

```python
df4_mean.isnull().sum().sum()
```

```python
for i ,var in enumerate(mising_num):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num [var],bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_mean[var],bins=20, kde_kws={'linewidth':5, 'color':'green'}, label="mean",)
    plt.legend()
    
```

```python
df5_median=df3_num.fillna(df3_num.median())
```

```python
df5_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
for i,var in enumerate(mising_num):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_mean[var],bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'}, label="mean",)
    sns.distplot(df5_median[var],bins=20,hist=False, kde_kws={'linewidth':3, 'color':'k'}, label="median",)
    plt.legend()
    

    
# plt.figure(figsize=(10,10))
# sns.set()
# for i, var in enumerate(missing_num_var):
#     plt.subplot(2,2,i+1)
#     sns.distplot(df3_num[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
#     sns.distplot(df4_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
#     sns.distplot(df5_num_median[var], bins=20,hist=False, kde_kws={'linewidth':3, 'color':'k'},label="Median",)
#     plt.legend()
    
```
