```python
df_copy=df.copy()
num_var_miss=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
num_cat=['LotConfig','MasVnrType','GarageType']
for var,cat in zip(num_var_miss,num_cat):
    for var1 in df[cat].unique():
        df_copy.update(df[df.loc[:,cat]==var1][var].replace(np.nan,df[df.loc[:,cat]==var1][var].mean()))
        
    
```

```python
df_copy[num_var_miss].isnull().sum()
```

```python
df_copy=df.copy()
num_var_miss=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
num_cat = ['LotConfig','Exterior2nd','KitchenQual']
for var,cat in zip(num_var_miss,num_cat):
    for var1 in df[cat].unique():
        df_copy.update(df[df.loc[:,cat]==var1][var].replace(np.nan,df[df.loc[:,cat]==var1][var].mean()))
```

```python
df_copy[num_var_miss].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_var_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df_copy[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    plt.legend()
```

```python
df_copy2=df.copy()
num_var_miss=['LotFrontage', 'MasVnrArea', 'GarageYrBlt']
num_cat = ['LotConfig','Exterior2nd','KitchenQual']
for var,cat in zip(num_var_miss,num_cat):
    for var1 in df[cat].unique():
        df_copy2.update(df[df.loc[:,cat]==var1][var].replace(np.nan,df[df.loc[:,cat]==var1][var].median()))
```

```python
df_copy2[num_var_miss].isnull().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(num_var_miss):
    plt.subplot(2,2,i+1)
    sns.distplot(df[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df_copy[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    sns.distplot(df_copy2[var], bins=20, kde_kws={'linewidth':3, 'color':'k'},label="Median")
    plt.legend()
```
